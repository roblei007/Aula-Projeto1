# 🗂️ CASO #1 — A Corrida Impossível
## *Paradoxo de Zenão de Eleia (séc. V a.C.)*

### 📜 Briefing

Recruta, este caso veio de **2.500 anos atrás**, mas continua atormentando filósofos.

**O cenário:** O herói **Aquiles**, o homem mais rápido da Grécia, vai apostar uma corrida com uma **tartaruga**. Sendo cavalheiro, Aquiles dá **100 metros de vantagem** à tartaruga. Aquiles corre a `10 m/s`, a tartaruga a `1 m/s`.

**A acusação de Zenão:** *Aquiles JAMAIS alcançará a tartaruga.*

### 🤔 O argumento (preste atenção!)

> Para Aquiles alcançar a tartaruga, primeiro ele tem que **chegar onde ela está agora** (100 m). Mas quando ele chega lá, a tartaruga já andou um pouco mais. Então ele tem que chegar a esse novo ponto — e quando chega, ela andou MAIS um pouco. E assim por diante, **infinitamente**. Logo, Aquiles nunca alcança a tartaruga.

Faz sentido? E no entanto sabemos que, na vida real, Aquiles passaria pela tartaruga em poucos segundos. **O que está errado no raciocínio de Zenão?**

### 💻 Evidência computacional

Vamos simular a corrida e ver o que acontece. Sua primeira tarefa é **rodar o código abaixo** e analisar a saída.

````python
# ============================================================
# CASO #1 — Simulação da corrida de Aquiles vs Tartaruga
# Vamos seguir a LÓGICA DE ZENÃO passo a passo
# ============================================================

# Posições iniciais (em metros)
posicao_aquiles = 0       # Aquiles começa na posição 0
posicao_tartaruga = 100   # Tartaruga começa 100 metros à frente

# Velocidades (metros por segundo)
velocidade_aquiles = 10   # 10 m/s — bem rápido
velocidade_tartaruga = 1  # 1 m/s — bem devagar

# Vamos seguir a lógica de Zenão por 10 'etapas'
# A cada etapa, Aquiles vai até onde a tartaruga ESTAVA
etapa = 1

# Loop que repete o raciocínio de Zenão 10 vezes
while etapa <= 10:
    # Calcula quanto tempo Aquiles leva para chegar à posição da tartaruga
    distancia = posicao_tartaruga - posicao_aquiles
    tempo = distancia / velocidade_aquiles

    # Atualiza as posições
    posicao_aquiles = posicao_tartaruga          # Aquiles chegou onde a tartaruga estava
    posicao_tartaruga += velocidade_tartaruga * tempo  # Tartaruga andou nesse tempo

    # Mostra a etapa formatada
    print(f"Etapa {etapa:2d}: Aquiles em {posicao_aquiles:.6f} m | Tartaruga em {posicao_tartaruga:.6f} m | Distância: {posicao_tartaruga - posicao_aquiles:.6f} m")

    etapa += 1

print("\n🤔 Observe: a distância NUNCA chega a zero, mas fica cada vez menor.")
print("   Isso é o que Zenão chamou de 'paradoxo'.")
````

### 🔍 A solução do mistério

**Aquiles ALCANÇA a tartaruga sim!** O erro de Zenão foi assumir que **infinitos passos = tempo infinito**. Mas se cada passo dura cada vez menos tempo, a soma de infinitos passos pode resultar num **tempo finito** (isso se chama "série convergente" na matemática).

Vamos calcular o **momento exato** em que Aquiles ultrapassa a tartaruga usando física simples:

````python
# ============================================================
# A SOLUÇÃO: cálculo direto do encontro
# ============================================================

# Equação: posição_aquiles(t) = posição_tartaruga(t)
# 0 + 10*t = 100 + 1*t
# 9*t = 100
# t = 100 / 9

vantagem = 100
v_aquiles = 10
v_tartaruga = 1

# Tempo do encontro (matemática direta, sem mistério)
tempo_encontro = vantagem / (v_aquiles - v_tartaruga)
posicao_encontro = v_aquiles * tempo_encontro

print(f"🏁 Aquiles alcança a tartaruga em {tempo_encontro:.4f} segundos")
print(f"📍 Na posição: {posicao_encontro:.4f} metros")
print(f"\n💡 A 'soma infinita' do Zenão converge para esse tempo finito!")
````

### 🎯 Sua missão #1

**Dra. Lovelace exige uma investigação mais profunda.** Crie uma função `simular_corrida(vantagem, v_aquiles, v_tartaruga, etapas)` que:

1. Recebe a vantagem inicial da tartaruga, as duas velocidades e o número de etapas a simular.
2. Usa um `for` com `range(etapas)` para simular a lógica de Zenão.
3. Retorna uma **lista** com a distância restante em cada etapa.
4. Imprima também o tempo real do encontro (cálculo direto).

**Teste sua função com 3 cenários diferentes:**
- Cenário A: vantagem=100, Aquiles=10, Tartaruga=1, 8 etapas
- Cenário B: vantagem=50, Aquiles=20, Tartaruga=2, 5 etapas
- Cenário C: vantagem=200, Aquiles=15, Tartaruga=3, 6 etapas

````python
# 👉 Sua investigação aqui:

def simular_corrida(vantagem, v_aquiles, v_tartaruga, etapas):
    # Lista que vai armazenar as distâncias
    distancias = []

    # Suas variáveis iniciais aqui
    posicao_aquiles = 0
    posicao_tartaruga = vantagem

    # Seu loop aqui (use for com range(etapas))
    for etapa_num in range(1, etapas + 1):
        # Calcula quanto tempo Aquiles leva para chegar à posição da tartaruga
        distancia_atual = posicao_tartaruga - posicao_aquiles
        if distancia_atual <= 0:
            # Aquiles já alcançou ou ultrapassou
            distancias.append(0.0)
            break
        tempo_passo = distancia_atual / v_aquiles

        # Atualiza as posições
        posicao_aquiles = posicao_tartaruga          # Aquiles chegou onde a tartaruga estava
        posicao_tartaruga += v_tartaruga * tempo_passo  # Tartaruga andou nesse tempo

        distancias.append(posicao_tartaruga - posicao_aquiles)

    # Calcule e mostre o tempo real do encontro (fora do loop, mas dentro da função)
    tempo_encontro_real = vantagem / (v_aquiles - v_tartaruga)
    posicao_encontro_real = v_aquiles * tempo_encontro_real

    print(f"🏁 Aquiles alcança a tartaruga em {tempo_encontro_real:.4f} segundos")
    print(f"📍 Na posição: {posicao_encontro_real:.4f} metros")
    print(f"\n💡 A 'soma infinita' do Zenão converge para esse tempo finito!")

    return distancias


# Teste os 3 cenários:

print("\n--- Cenário A ---")
# Cenário A
vantagem_a = 100
v_aquiles_a = 10
v_tartaruga_a = 1
etapas_a = 8
dist_a = simular_corrida(vantagem_a, v_aquiles_a, v_tartaruga_a, etapas_a)
print(f"Distâncias restantes no Cenário A: {dist_a}")

print("\n--- Cenário B ---")
# Cenário B
vantagem_b = 50
v_aquiles_b = 20
v_tartaruga_b = 2
etapas_b = 5
dist_b = simular_corrida(vantagem_b, v_aquiles_b, v_tartaruga_b, etapas_b)
print(f"Distâncias restantes no Cenário B: {dist_b}")

print("\n--- Cenário C ---")
# Cenário C
vantagem_c = 200
v_aquiles_c = 15
v_tartaruga_c = 3
etapas_c = 6
dist_c = simular_corrida(vantagem_c, v_aquiles_c, v_tartaruga_c, etapas_c)
print(f"Distâncias restantes no Cenário C: {dist_c}")
````

# 🗂️ CASO #2 — O Hotel Sem Vagas… que Sempre Tem Vaga
## *Paradoxo do Hotel de Hilbert (David Hilbert, 1924)*

### 📜 Briefing

Recruta, este é um caso peculiar. Recebemos uma denúncia de um hotel muito... incomum.

**O cenário:** O **Hotel Infinito** tem **infinitos quartos** (1, 2, 3, 4, ... sem fim), e está com **TODOS** os quartos ocupados. Lotação máxima. Não há nem uma vaga.

Aí chega um novo hóspede pedindo um quarto. O recepcionista, sem hesitar, diz: **"Claro! Temos vaga."**

**Como é possível?** 🤯

### 🤔 A solução engenhosa de Hilbert

O recepcionista pega o microfone e anuncia:

> *"Atenção, hóspedes! Por favor, cada um de vocês se mude para o quarto com o número seguinte ao seu atual. Quem está no 1, vai pro 2. Quem está no 2, vai pro 3. E assim por diante."*

Resultado: o **quarto 1 fica livre** para o novo hóspede! E ninguém ficou de fora, porque sempre há um "próximo quarto". É a magia do **infinito**.

### 🆕 Nova ferramenta: `set` (conjuntos)

Para investigar este caso, vamos aprender uma nova estrutura: o **set**.

**Características do `set`:**
- Criado com `{ }` ou `set()`
- **Não permite duplicatas** (perfeito para uma lista de quartos!)
- **Não tem ordem** (não há índices)
- Operações de conjunto: união (`|`), interseção (`&`), diferença (`-`)
- Métodos: `.add()`, `.remove()`, `.discard()`

**Diferença para listas:**

| Lista `[]` | Set `{}` |
|------------|----------|
| Aceita duplicatas | Não aceita duplicatas |
| Tem ordem | Sem ordem |
| Acesso por índice | Sem índice |
| Mais lenta para `in` | **Muito mais rápida** para `in` |

````python
# ============================================================
# Aprendendo sets antes de resolver o caso
# ============================================================

# Criando um set com chaves {}
frutas = {"maçã", "banana", "uva", "banana", "maçã"}
print("Set de frutas:", frutas)   # Note: 'banana' e 'maçã' aparecem só UMA vez!

# Adicionar item
frutas.add("laranja")
print("Após adicionar laranja:", frutas)

# Adicionar item duplicado: o set ignora
frutas.add("maçã")   # já existe, nada acontece
print("Após tentar adicionar maçã de novo:", frutas)

# Remover (atenção: .remove() dá erro se não existir; .discard() não)
frutas.discard("banana")
frutas.discard("kiwi")   # não existe, mas não dá erro
print("Após remover banana:", frutas)

# Verificar pertinência (super rápido em sets!)
print("'uva' está? ", "uva" in frutas)   # True
print("'pera' está?", "pera" in frutas)  # False
````

````python
# ============================================================
# Operações de conjunto: união, interseção, diferença
# ============================================================

# Imagine duas turmas de uma escola
turma_python = {"Ana", "Bruno", "Carlos", "Diana"}
turma_javascript = {"Carlos", "Diana", "Eduardo", "Fernanda"}

# UNIÃO (|): todos os alunos de qualquer turma
todos = turma_python | turma_javascript
print("Todos os alunos:", todos)

# INTERSEÇÃO (&): alunos que estão nas DUAS turmas
ambas = turma_python & turma_javascript
print("Em ambas as turmas:", ambas)

# DIFERENÇA (-): só Python, sem JavaScript
so_python = turma_python - turma_javascript
print("Só fazem Python:", so_python)

# DIFERENÇA SIMÉTRICA (^): em uma OU outra, mas não em ambas
uma_ou_outra = turma_python ^ turma_javascript
print("Uma ou outra (não ambas):", uma_ou_outra)
````

### 💻 Evidência computacional do Hotel de Hilbert

Não conseguimos simular *infinito* literalmente, mas podemos simular um **hotel muito grande** (1 milhão de quartos) e mostrar que a manobra do Hilbert funciona.

````python
# ============================================================
# Simulação do Hotel de Hilbert (versão finita, mas grande)
# ============================================================

# O hotel tem 1 milhão de quartos, todos ocupados
# Vamos representar os quartos ocupados com um set
tamanho_hotel = 1_000_000   # underscores no número facilitam a leitura

# Cria um set com todos os quartos numerados de 1 até 1.000.000
# (em sets reais, o range cria um conjunto eficientemente)
quartos_ocupados = set(range(1, tamanho_hotel + 1))

print(f"🏨 Hotel inicial: {len(quartos_ocupados)} quartos ocupados")
print(f"   Quarto 1 está ocupado? {1 in quartos_ocupados}")
print(f"   Quarto 999.999 está ocupado? {999_999 in quartos_ocupados}")

# === A MANOBRA DE HILBERT ===
# Cada hóspede do quarto N se muda para o quarto N+1
# Isso é o que o recepcionista anuncia no microfone
novos_quartos = set()
for quarto in quartos_ocupados:
    novos_quartos.add(quarto + 1)   # cada hóspede vai para o próximo

quartos_ocupados = novos_quartos

# Verificando o resultado
print(f"\n🔄 Após a manobra de Hilbert:")
print(f"   Quarto 1 está ocupado? {1 in quartos_ocupados}")   # ✨ FALSE!
print(f"   Quarto 2 está ocupado? {2 in quartos_ocupados}")
print(f"   Total de hóspedes: {len(quartos_ocupados)} (mesmo número, todos acomodados!)")

# 🎉 O quarto 1 está livre para o novo hóspede!
quartos_ocupados.add(1)
print(f"\n✅ Novo hóspede acomodado no quarto 1.")
print(f"   Total agora: {len(quartos_ocupados)} hóspedes.")
````

Chegou um **ônibus com infinitos passageiros** querendo se hospedar. O recepcionista resolve assim: cada hóspede atual do quarto **N** se muda para o quarto **2N** (1→2, 2→4, 3→6, ...). Isso libera **todos os quartos ímpares** para os novos passageiros do ônibus!

Sua missão:

4. Use **operações de conjunto** para descobrir quais quartos ímpares de 1 a 200 estão livres (use `set(range(...))` e o operador `-`).

````python
# 👉 Sua investigação aqui:

# 1. Set inicial de quartos ocupados (1 a 100)
quartos_ocupados = set(range(1, 100))

# 2. Função manobra_pares
def manobra_pares(quartos):
    novos_quartos = set()
    for quarto in quartos:

         # Cada hóspede N vai para o quarto 2N
        novos_quartos.add(quarto * 2)
    return novos_quartos

# 3. Aplique a manobra e verifique
quartos_depois_da_manobra = manobra_pares(quartos_ocupados)

print(f"🏨 Quartos ocupados após a manobra: {len(quartos_depois_da_manobra)}")
print(f"Quarto 1 está livre? {1 not in quartos_depois_da_manobra}")
print(f"Quarto 2 está livre? {2 not in quartos_depois_da_manobra}")
print(f"Quarto 100 está ocupado? {100 in quartos_depois_da_manobra}")
print(f"Quarto 200 está ocupado? {200 in quartos_depois_da_manobra}")

# 4. Usando operações de conjunto, ache os ímpares livres de 1 a 200
todos_impares = set(range(1, 200, 2))
impares_livres = todos_impares - quartos_depois_da_manobra

print(f"\n✨ Total de quartos ímpares livres (1-200): {len(impares_livres)}")
print(f"Alguns quartos ímpares livres: {sorted(list(impares_livres))[:10]}...")
````

# 🗂️ CASO #3 — A Porta da Fortuna
## *Paradoxo de Monty Hall (1975)*

### 📜 Briefing

**ATENÇÃO, RECRUTA!** Este é o caso que **mais divide opiniões** na história da estatística. Quando foi publicado em 1990, **mais de 1000 PhDs em matemática escreveram cartas dizendo que a resposta estava errada**. Estavam todos enganados.

**O cenário:** Você está num programa de TV. À sua frente, três portas:
- 🚪 Atrás de **uma** delas: um carro 🚗
- 🚪 Atrás das **outras duas**: bodes 🐐

Você escolhe uma porta (por exemplo, a Porta 1).

Aí o apresentador, **Monty Hall**, que sabe onde está o carro, **abre uma das outras duas portas mostrando um bode** (digamos, a Porta 3).

Agora ele te oferece: **"Quer trocar para a Porta 2 ou ficar com a Porta 1?"**

### 🤔 A pergunta dos R$ 1.000.000

**Vale a pena trocar de porta?**

A intuição diz: "Tanto faz, é 50/50, sobraram duas portas."

**A intuição está ERRADA.**

**Trocar dobra suas chances** de ganhar! Quem fica: 1/3 (33%). Quem troca: 2/3 (66%). 🤯

### 🆕 Nova ferramenta: módulo `random`

Para provar isso, vamos rodar a simulação **10.000 vezes**. E aí precisamos do módulo `random` do Python.

**Funções principais do `random`:**
- `random.randint(a, b)` — número inteiro aleatório entre `a` e `b` (incluindo os dois)
- `random.choice(lista)` — sorteia um item de uma lista
- `random.shuffle(lista)` — embaralha a lista no lugar
- `random.random()` — número decimal aleatório entre 0 e 1
- `random.sample(lista, n)` — sorteia `n` itens distintos

Para usar, primeiro **importamos** o módulo:

````python
# ============================================================
# Conhecendo o módulo random
# ============================================================

# Importamos o módulo no topo do programa
import random

# randint(a, b): número inteiro entre a e b (inclusive os dois)
dado = random.randint(1, 6)   # simula um dado
print(f"Dado: {dado}")

# choice(lista): pega um item aleatório da lista
frutas = ["maçã", "banana", "uva", "manga"]
escolhida = random.choice(frutas)
print(f"Fruta sorteada: {escolhida}")

# shuffle(lista): embaralha a lista (modifica a original!)
baralho = ["A", "K", "Q", "J", "10"]
random.shuffle(baralho)
print(f"Baralho embaralhado: {baralho}")

# random(): número decimal entre 0 e 1
probabilidade = random.random()
print(f"Número entre 0 e 1: {probabilidade:.4f}")

# sample(lista, n): n itens distintos da lista
loteria = random.sample(range(1, 61), 6)   # 6 números da Mega-Sena
print(f"Mega-Sena: {sorted(loteria)}")
````

### 💻 Evidência computacional: simulação do Monty Hall

Vamos rodar o jogo **10.000 vezes** com duas estratégias e ver os resultados.

**Estratégia A:** Sempre fica com a porta original.
**Estratégia B:** Sempre troca de porta.

Se a intuição ("50/50") estiver certa, as duas estratégias devem ganhar mais ou menos a mesma quantidade de vezes. Vamos ver...

````python
# ============================================================
# Simulação: Monty Hall — vale a pena trocar de porta?
# ============================================================

import random

def jogar_monty_hall(trocar_de_porta):
    """
    Simula UMA rodada do jogo. Retorna True se ganhou o carro, False se ganhou bode.
    Parâmetro trocar_de_porta: True se a estratégia é trocar, False se é ficar.
    """
    # As 3 portas (numeradas 0, 1, 2)
    portas = [0, 1, 2]

    # Sorteia onde está o carro
    porta_do_carro = random.choice(portas)

    # O jogador escolhe uma porta aleatória
    escolha_inicial = random.choice(portas)

    # Monty abre uma porta que: (1) não é a do jogador (2) não tem o carro
    # Lista de portas que Monty PODE abrir
    portas_que_monty_pode_abrir = []
    for porta in portas:
        if porta != escolha_inicial and porta != porta_do_carro:
            portas_que_monty_pode_abrir.append(porta)

    # Monty escolhe uma dessas portas (pode ser uma ou duas opções)
    porta_aberta_por_monty = random.choice(portas_que_monty_pode_abrir)

    # Decide a escolha final do jogador
    if trocar_de_porta:
        # Troca para a porta que sobrou (não é a inicial e não foi aberta)
        for porta in portas:
            if porta != escolha_inicial and porta != porta_aberta_por_monty:
                escolha_final = porta
                break
    else:
        # Fica com a escolha inicial
        escolha_final = escolha_inicial

    # Retorna True se ganhou o carro
    return escolha_final == porta_do_carro


# === RODANDO A SIMULAÇÃO ===
rodadas = 10_000

# Estratégia A: NÃO trocar
vitorias_ficando = 0
for _ in range(rodadas):
    if jogar_monty_hall(trocar_de_porta=False):
        vitorias_ficando += 1

# Estratégia B: SEMPRE trocar
vitorias_trocando = 0
for _ in range(rodadas):
    if jogar_monty_hall(trocar_de_porta=True):
        vitorias_trocando += 1

# Resultados
print(f"📊 RESULTADOS APÓS {rodadas:,} RODADAS:")
print(f"\n🚪 Estratégia FICAR:   {vitorias_ficando:,} vitórias ({vitorias_ficando/rodadas*100:.2f}%)")
print(f"🔄 Estratégia TROCAR:  {vitorias_trocando:,} vitórias ({vitorias_trocando/rodadas*100:.2f}%)")
print(f"\n💡 A intuição '50/50' está errada. Trocar GANHA o dobro!")
````

### 🔍 Por que trocar é melhor?

A explicação intuitiva: quando você escolheu a porta inicial, tinha **1/3 de chance de acertar**. Logo, a chance do carro estar nas **outras duas portas** é **2/3**.

Quando Monty abre uma porta com bode, ele **concentra esses 2/3 numa única porta restante**. Por isso trocar dobra suas chances!

### 🎯 Sua missão #3

Vamos investigar uma **variação** do paradoxo. Crie uma função `monty_hall_n_portas(n_portas, trocar)` que generaliza o jogo para **N portas** (em vez de só 3):

1. A função recebe o número de portas e se vai trocar.
2. O carro é colocado atrás de uma porta aleatória.
3. O jogador escolhe uma porta aleatória.
4. Monty abre **TODAS as portas restantes que têm bode**, exceto uma (sempre sobram exatamente 2 portas no final).
5. Se `trocar=True`, vai para a porta restante.
6. Retorna `True` se ganhou.

**Depois:** Rode 10.000 simulações para 3, 5, 10 e 100 portas, comparando as estratégias. **Você vai se assustar com o resultado para 100 portas!** 🤯

````python
# 👉 Sua investigação aqui:

import random

def monty_hall_n_portas(n_portas, trocar):
    # 1. Cria a lista de portas
    portas = list(range(n_portas))

    # 2. Sorteia onde está o carro
    porta_do_carro = random.choice(portas)

    # 3. Jogador escolhe uma porta aleatória
    escolha_inicial = random.choice(portas)

    # 4. Monty abre todas as portas com bode, exceto uma
    # Monty deixa fechadas: a escolha do jogador e mais UMA que obrigatoriamente contém o carro
    # (ou uma porta aleatória com bode se o jogador já escolheu o carro)
    if escolha_inicial == porta_do_carro:
        # Jogador acertou de primeira, Monty deixa qualquer outra fechada
        opcoes_para_manter_fechada = [p for p in portas if p != escolha_inicial]
        outra_porta_fechada = random.choice(opcoes_para_manter_fechada)
    else:
        # Jogador errou, Monty é obrigado a manter a porta do carro fechada
        outra_porta_fechada = porta_do_carro

    # 5. Se trocar, escolha final é a outra porta que ficou fechada
    if trocar:
        escolha_final = outra_porta_fechada
    else:
        escolha_final = escolha_inicial

    # 6. Retorna True se ganhou
    return escolha_final == porta_do_carro

# === Teste com diferentes números de portas ===
rodadas = 10_000

for n in [3, 5, 10, 100]:
    vitorias_ficar = 0
    vitorias_trocar = 0

    for _ in range(rodadas):
        if monty_hall_n_portas(n, trocar=False): vitorias_ficar += 1
        if monty_hall_n_portas(n, trocar=True): vitorias_trocar += 1

    prob_ficar = (vitorias_ficar / rodadas) * 100
    prob_trocar = (vitorias_trocar / rodadas) * 100

    print(f"Com {n:3} portas — Ficar: {prob_ficar:5.2f}% | Trocar: {prob_trocar:5.2f}%")
````

# 🗂️ CASO #4 — A Festa dos Coincidentes
## *Paradoxo do Aniversário*

### 📜 Briefing

Recruta, este caso vai te fazer duvidar de festas pra sempre.

**A pergunta:** Numa sala com **23 pessoas**, qual a probabilidade de **pelo menos duas** fazerem aniversário no mesmo dia?

**A intuição** diz: "Tem 365 dias no ano e só 23 pessoas... deve ser uns 5%, no máximo."

**A realidade:** É **mais de 50%**. Com 23 pessoas, é **mais provável que haja coincidência do que não haja**. 🎂🎂

Com **70 pessoas**, a probabilidade passa de **99,9%**.

### 🤔 Por que a intuição falha?

A gente tende a pensar "qual a chance de alguém fazer aniversário no MESMO DIA QUE EU?" Mas a pergunta é diferente: **qual a chance de QUALQUER PAR de pessoas coincidir?**

Com 23 pessoas, há **253 pares possíveis** de comparação (a fórmula é `n*(n-1)/2`). É muita oportunidade de coincidência!

### 💻 Evidência computacional

````python
# ============================================================
# Simulação do Paradoxo do Aniversário
# ============================================================

import random

def tem_coincidencia(n_pessoas):
    """
    Simula uma sala com n_pessoas e retorna True se há ao menos
    duas com o mesmo aniversário.
    """
    # Lista que vai guardar os aniversários sorteados (1 a 365)
    aniversarios = []

    for _ in range(n_pessoas):
        # Sorteia um dia do ano
        dia = random.randint(1, 365)

        # Se o dia já apareceu, achamos uma coincidência!
        if dia in aniversarios:
            return True
        aniversarios.append(dia)

    # Se chegou aqui, não houve coincidência
    return False


# === Rodando a simulação para diferentes tamanhos de sala ===
rodadas = 10_000   # rodamos 10 mil simulações para cada tamanho

# Dicionário para guardar os resultados (aplicação direta de dicionário!)
resultados = {}

# Tamanhos de sala que vamos testar
tamanhos = [5, 10, 15, 20, 23, 30, 40, 50, 70, 100]

for n in tamanhos:
    coincidencias = 0
    for _ in range(rodadas):
        if tem_coincidencia(n):
            coincidencias += 1

    probabilidade = coincidencias / rodadas * 100
    resultados[n] = probabilidade

# Mostra os resultados em formato de tabela
print("📊 PROBABILIDADE DE COINCIDÊNCIA DE ANIVERSÁRIO")
print("=" * 50)
print(f"{'Pessoas':>10} | {'Probabilidade':>15} | Visualização")
print("-" * 50)
for n, prob in resultados.items():
    # Cria uma 'barra' visual proporcional à probabilidade
    barra = "█" * int(prob / 2)   # cada bloco = 2%
    print(f"{n:>10} | {prob:>13.2f}% | {barra}")

print("\n💡 Observe: com apenas 23 pessoas, a probabilidade já passa de 50%!")
````

````python
# ============================================================
# BÔNUS: descobrindo QUAIS dias coincidiram (com dicionário)
# ============================================================

import random

def festa_com_detalhes(n_pessoas):
    """
    Simula uma festa e retorna um dicionário com:
    - 'aniversariantes': dicionário {dia: lista de pessoas}
    - 'coincidencias': lista de dias com mais de uma pessoa
    """
    aniversariantes = {}

    for i in range(1, n_pessoas + 1):
        nome = f"Pessoa {i}"
        dia = random.randint(1, 365)

        # Se o dia já existe no dicionário, adiciona à lista
        # Se não, cria a chave com uma lista contendo essa pessoa
        if dia in aniversariantes:
            aniversariantes[dia].append(nome)
        else:
            aniversariantes[dia] = [nome]

    # Filtra apenas os dias com coincidências
    coincidencias = {}
    for dia, pessoas in aniversariantes.items():
        if len(pessoas) > 1:
            coincidencias[dia] = pessoas

    return aniversariantes, coincidencias


# Simula uma festa com 30 pessoas
aniversariantes, coincidencias = festa_com_detalhes(30)

print(f"🎉 Festa com 30 pessoas — {len(aniversariantes)} dias diferentes\n")

if coincidencias:
    print("🎂 Coincidências encontradas:")
    for dia, pessoas in coincidencias.items():
        # Junta os nomes com vírgula usando o método .join()
        nomes = ", ".join(pessoas)
        print(f"   Dia {dia}: {nomes}")
else:
    print("😴 Nenhuma coincidência nesta festa (vai acontecer em breve!)")
````

# 🗂️ CASO #5 — O Cretense que Mente
## *Paradoxo do Mentiroso (Epimênides, séc. VI a.C.)*

### 📜 Briefing

Recruta, este caso é **diferente** dos anteriores. Aqui não tem como simular: a lógica simplesmente **quebra**.

**A frase em questão:**

> *"Esta frase é falsa."*

**Analise:** Se a frase é **verdadeira**, então o que ela diz é verdade — ou seja, ela é **falsa**. Mas se é falsa, então o que ela diz é falso — ou seja, ela é **verdadeira**. 🌀

É um loop infinito de contradição. Em filosofia, isso é chamado de **autorreferência paradoxal**.

### 🤔 Por que isso importa em programação?

Porque **programas podem entrar em situações paradoxais**: divisão por zero, acesso a listas com índice errado, conversões impossíveis... Quando isso acontece, o Python **levanta uma exceção** (um "erro"). Se você não tratar, o programa **TRAVA**.

É aqui que entra nossa nova ferramenta.

### 🆕 Nova ferramenta: `try` / `except`

**Sintaxe:**
```python
try:
    # código que pode dar erro
except TipoDeErro:
    # o que fazer SE o erro acontecer
```

É como um **detector de paradoxos**: tentamos executar algo arriscado e, se a lógica quebrar, capturamos o erro em vez de deixar o programa travar.

**Tipos comuns de erro:**
- `ZeroDivisionError` — divisão por zero
- `ValueError` — valor inválido (ex: converter "abc" para int)
- `TypeError` — tipo incompatível (ex: somar string com número)
- `KeyError` — chave que não existe num dicionário
- `IndexError` — índice fora do alcance da lista
- `FileNotFoundError` — arquivo não encontrado

````python
# ============================================================
# Lidando com 'paradoxos' (erros) no Python
# ============================================================

# === SEM try/except: o programa TRAVA ===
# Descomente a linha abaixo para ver o erro:
# resultado = 10 / 0   # ZeroDivisionError

# === COM try/except: capturamos o erro ===
try:
    resultado = 10 / 0
    print(f"Resultado: {resultado}")
except ZeroDivisionError:
    print("❌ Detectado: tentativa de divisão por zero!")
    print("✅ Programa segue normalmente.")

print("\nO programa não travou. Continuamos investigando...")
````

````python
# ============================================================
# Capturando MÚLTIPLOS tipos de erro
# ============================================================

# Função para converter texto em número, com proteção
def converter_para_numero(texto):
    try:
        # Tenta converter
        numero = int(texto)
        return numero
    except ValueError:
        # Se o texto não for um número válido
        print(f"⚠️ '{texto}' não é um número válido.")
        return None

# Testando com vários casos
casos = ["42", "abc", "3.14", "-7", "", "100"]
for caso in casos:
    resultado = converter_para_numero(caso)
    print(f"   Entrada: '{caso}' → Resultado: {resultado}\n")
````

````python
# ============================================================
# try / except / else / finally
# ============================================================

# Estrutura completa do tratamento de erros:
# - try: o código arriscado
# - except: o que fazer se der erro
# - else: o que fazer SE NÃO deu erro
# - finally: SEMPRE executa, com erro ou sem erro

def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("❌ Não posso dividir por zero!")
        return None
    except TypeError:
        print("❌ Os dois valores precisam ser números!")
        return None
    else:
        # Executa só se NÃO houve erro
        print(f"✅ Divisão bem-sucedida: {resultado}")
        return resultado
    finally:
        # SEMPRE executa, mesmo que o programa esteja em chamas 🔥
        print("📋 Operação registrada no log.\n")

# Testes
dividir(10, 2)         # caso normal
dividir(10, 0)         # divisão por zero
dividir(10, "abc")     # tipo errado
````

````python
# ============================================================
# Aplicação prática: um input à prova de paradoxos
# ============================================================

# Função que pede um número ao usuário, repetindo até receber algo válido
def pedir_numero(mensagem):
    while True:   # repete até dar 'return' (o break implícito)
        try:
            valor = float(input(mensagem))
            return valor   # se chegou aqui, é número válido
        except ValueError:
            print("⚠️ Por favor, digite um número válido.")

# Teste (descomente para usar com input):
# idade = pedir_numero("Digite sua idade: ")
# print(f"Sua idade: {idade}")

print("💡 Esse padrão (try/except dentro de while True) é MUITO usado em programas reais!")
````

Recruta,

Os 5 casos foram só **treinamento**. O verdadeiro caso da DIP é o seguinte:

Um cassino misterioso oferece o seguinte jogo: você paga R$ 10 para entrar. Depois, você joga um **dado de 6 faces**. Se sair **6**, você ganha R$ 50. Se sair qualquer outra coisa, você não ganha nada.

**Pergunta:** vale a pena jogar?

Sua intuição talvez diga: "6/6 chance de perder R$ 10 e 1/6 de ganhar R$ 50... parece justo?"

**Sua missão final é DESCOBRIR a verdade através de simulação.**

### 🎯 Suas tarefas

Construa um programa que use **TUDO o que aprendeu**:

1. **Função** `jogar_uma_partida()` que:
   - Sorteia um dado (1 a 6) usando `random.randint`.
   - Retorna o **lucro líquido** dessa partida (50 - 10 = 40 se ganhar; -10 se perder).

2. **Função** `simular_jogos(n_partidas)` que:
   - Roda `n_partidas` partidas.
   - Retorna um **dicionário** com:
     - `"total_ganho"`: lucro acumulado
     - `"vitorias"`: número de vitórias
     - `"derrotas"`: número de derrotas
     - `"historico"`: **set** com os resultados únicos dos dados (1 a 6) que apareceram

3. **Função** `pedir_inteiro_seguro(mensagem)` que:
   - Usa `try/except` para garantir que o usuário digite um inteiro positivo.

4. **Programa principal** que:
   - Pergunta ao usuário quantas partidas quer simular (use a função segura).
   - Roda a simulação.
   - Mostra um relatório com:
     - Total de partidas, vitórias, derrotas
     - Lucro/prejuízo total
     - Lucro/prejuízo médio por partida
     - Quais faces do dado apareceram (do `set`)
     - **Veredito**: vale a pena jogar? (use um `if` baseado no lucro médio)

5. **Reflexão final** (em comentário): com base no resultado, qual a **expectativa matemática** desse jogo? *(Dica: probabilidade de ganhar × valor + probabilidade de perder × prejuízo)*

````python
# 👉 SEU CASO FINAL AQUI:

import random

# 1. Função jogar_uma_partida()
def jogar_uma_partida():
    dado = random.randint(1, 6)
    if dado == 6:
        return 40, dado  # Ganhou (50 - 10)
    else:
        return -10, dado # Perdeu (-10)

# 2. Função simular_jogos(n_partidas)
def simular_jogos(n_partidas):
    total_ganho = 0
    vitorias = 0
    derrotas = 0
    historico_faces = set()

    for _ in range(n_partidas):
        lucro, face = jogar_uma_partida()
        total_ganho += lucro
        historico_faces.add(face)
        if lucro > 0:
            vitorias += 1
        else:
            derrotas += 1

    return {
        "total_ganho": total_ganho,
        "vitorias": vitorias,
        "derrotas": derrotas,
        "historico": historico_faces
    }

# 3. Função pedir_inteiro_seguro(mensagem)
def pedir_inteiro_seguro(mensagem):
    while True:
        try:
            valor = int(input(mensagem))
            if valor > 0:
                return valor
            print("⚠️ Por favor, digite um número maior que zero.")
        except ValueError:
            print("⚠️ Entrada inválida! Digite um número inteiro.")

# 4. Programa principal
print("🕵️ Investigação do Cassino Paradoxal")
n = pedir_inteiro_seguro("Quantas partidas quer simular? ")
res = simular_jogos(n)

lucro_medio = res['total_ganho'] / n

print(f"\n--- RELATÓRIO ---")
print(f"Total de partidas: {n}")
print(f"Vitórias: {res['vitorias']} | Derrotas: {res['derrotas']}")
print(f"Lucro/Prejuízo Total: R$ {res['total_ganho']}")
print(f"Lucro Médio por partida: R$ {lucro_medio:.2f}")
print(f"Faces que apareceram: {sorted(list(res['historico']))}")

if lucro_medio > 0:
    print("\n✅ Veredito: Vale a pena jogar!")
else:
    print("\n❌ Veredito: Não vale a pena. A casa sempre ganha!")

# 5. Reflexão final (comentário):
# Expectativa matemática:
# E = (1/6) * (+40) + (5/6) * (-10)
# E = (40/6) - (50/6) = -10/6 = -R$ 1.66 por partida.
# Conclusão: Matematicamente, o jogador perde em média R$ 1,66 a cada rodada.
````
