## 3. Dicionários (`dict`)

Um **dicionário** é uma estrutura de dados que armazena pares **chave → valor**.

📌 Você pode imaginar como:
- Uma ficha de cadastro (campo → valor)
- Um mini-banco de dados na memória

### 3.1 Criando um dicionário


```python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia de Software"
}

print(aluno)
print("Nome:", aluno["nome"])
print("Age:", aluno["idade"])
print(aluno["idade"], aluno["nome"])
```

### 3.2 Adicionando, alterando e removendo chaves


```python
aluno["nota_final"] = 8.5  # adicionando nova chave
print(aluno)

aluno["idade"] = 20  # alterando valor
print(aluno)

del aluno["curso"]  # removendo chave
print(aluno)
```

### 3.3 Métodos úteis de dicionário


```python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia de Software",
    "nota_final": 8.5
}

print("Chaves:", aluno.keys())
print("Valores:", aluno.values())
print("Itens (pares chave-valor):", aluno.items())
```

## 4. Estruturas condicionais: `if`, `elif`, `else`

As estruturas condicionais permitem que o programa **tome decisões**.

### 4.1 Exemplo simples


```python
numero = int(input("Digite um número: "))

if numero > 0:
    print("Positivo!")
elif numero < 0:
    print("Negativo!")
else:
    print("Zero!")
```

### 4.2 Par ou ímpar
Vamos usar o operador `%` (resto da divisão) para decidir se um número é par ou ímpar.

```python
num = int(input("Digite um número: "))

if num % 2 == 0:
    print("Par")
else:
    print("Ímpar")
```

## 5. Combinando dicionários com condicionais

Agora vamos usar **dicionários** junto com **if / elif / else** para tomar decisões.

```python
notas = {"Alice": 7.5, 'Bruno': 5.0, "Carla": 9.2}

aluno = input("Nome do aluno: ")

if aluno in notas:
    nota = notas[aluno]
    if nota >= 7:
        print(f"{aluno} foi APROVADO com nota {nota}")
    elif nota >= 5:
        print(f"{aluno} está em RECUPERAÇÃO com nota {nota}")
    else:
        print(f"{aluno} foi REPROVADO com nota {nota}")
else:
    print(f"{aluno} não está no sistema.")
```

```python
def soma(a, b, c):
    resultado = a + b + c
    return resultado

x = soma(3, 5, 2)
y = soma (1, 2, 3)
marcos = soma (8, 9, 10)

print("Resultado da soma:", x)
print("Resultado da soma:", y)
print("Resultado da soma:", marcos)
```

## 7. Desafio guiado — Mini sistema de cadastro de alunos

Vamos montar, passo a passo, um mini sistema em Python que:

1. Usa um **dicionário** para guardar alunos e suas notas
2. Usa **funções** para organizar o código
3. Usa **if / elif / else** para tomar decisões
4. Oferece um **menu** para o usuário escolher ações

### Passo 1 — Estrutura básica do dicionário


```python
alunos = {}
print("Sistema iniciado. Dicionário de alunos vazio:", alunos)
```

### Passo 2 — Função para adicionar aluno


```python
def adicionar_aluno(alunos):
    nome = input("Nome do aluno: ")
    nota = float(input("Nota do aluno: "))
    alunos[nome] = nota
    print(f"Aluno {nome} adicionado com nota {nota}.")

# Teste rápido
#adicionar_aluno(alunos)
#print(alunos)
```

### Passo 3 — Função para listar alunos


```python
def listar_alunos(alunos):
    if not alunos:
        print("Nenhum aluno cadastrado ainda.")
    else:
        print("\nLista de alunos:")
        for nome, nota in alunos.items():
            print(f"- {nome}: {nota}")

# Teste rápido
# listar_alunos(alunos)
```

### Passo 4 — Menu principal


```python
def menu():
    print("\n=== Sistema de Alunos ===")
    print("1 - Adicionar aluno")
    print("2 - Listar alunos")
    print("3 - Sair")

alunos = {}

while True:
    menu()
    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        adicionar_aluno(alunos)
    elif opcao == "2":
        listar_alunos(alunos)
    elif opcao == "3":
        print("Saindo do sistema...")
        break
    else:
        print("Opção inválida. Tente novamente.")
```

```python
estoque = {
    "Televisão": {"quantidade": 10, "preco": 2000.00},
    "Geladeira": {"quantidade": 4,  "preco": 3000.00},
    "Forno":     {"quantidade": 2,  "preco": 1000.00}
}

def valorTotalEstoque(estoque):
    total = 0
    for produto, dados in estoque.items():
        quantidade = dados["quantidade"]
        preco = dados["preco"]
        total += quantidade * preco
    return total

def avisaEstoqueBaixoRetorno(estoque):
    avisos = []
    for produto, dados in estoque.items():
        quantidade = dados["quantidade"]
        if quantidade < 5:
            avisos.append(f"{produto} está com baixo estoque! Quantidade: {quantidade}")
    return avisos

avisos = avisaEstoqueBaixoRetorno(estoque)

for aviso in avisos:
    print(aviso)


print("Valor total do estoque:", valorTotalEstoque(estoque))
avisaEstoqueBaixo(estoque)
```
