## Exercício 1: Classificar nível de experiência

Escreva uma função `classificar_nivel(anos)` que retorna o nível de um profissional a partir dos anos de experiência:

- 0 a 2 anos: `"Júnior"`
- 3 a 5 anos: `"Pleno"`
- 6 anos ou mais: `"Sênior"`
- Valor negativo: `"Inválido"`

**Exemplo:**

```
classificar_nivel(1)   ->  "Júnior"
classificar_nivel(4)   ->  "Pleno"
classificar_nivel(10)  ->  "Sênior"
classificar_nivel(-2)  ->  "Inválido"
```

````python
def classificar_nivel(anos):
    nivel = "Júnior"
    if anos >=0 and anos <=2:
      nivel = "Júnior"
    elif anos > 2 and anos <=5:
      nivel = "Pleno"
    elif anos >= 6:
      nivel = "Sênior"
    else:
      nivel = "Inválido"

    return nivel
````

````python
# Testes
print(classificar_nivel(0)  == "Júnior")
print(classificar_nivel(2)  == "Júnior")
print(classificar_nivel(3)  == "Pleno")
print(classificar_nivel(5)  == "Pleno")
print(classificar_nivel(6)  == "Sênior")
print(classificar_nivel(20) == "Sênior")
print(classificar_nivel(-1) == "Inválido")
````


## Exercício 2: Resumo de salários

Escreva uma função `resumo_salarios(salarios)` que recebe uma lista de números (salários anuais em USD) e retorna um dicionário com:

- `"contagem"`: quantos salários foram submetidos
- `"media"`: média dos salários, arredondada para 2 casas decimais
- `"minimo"`: menor valor
- `"maximo"`: maior valor


**Exemplo:**

```
resumo_salarios([50000, 80000, 120000, 95000, 60000])
->  {"contagem": 5, "media": 81000.0, "minimo": 50000, "maximo": 120000}
```


````python
def resumo_salarios(salarios):
    contagem = len(salarios)
    media = round(sum(salarios)/len(salarios),2)
    minimo = min(salarios)
    maximo = max(salarios)

    resultado = {"contagem":contagem,
                 "media":media,
                 "minimo":minimo,
                 "maximo":maximo}

    return resultado
````

````python
# Testes
r = resumo_salarios([50000, 80000, 120000, 95000, 60000])
print(r["contagem"]  == 5)
print(r["media"]  == 81000.0)
print(r["minimo"] == 50000)
print(r["maximo"] == 120000)
````


## **Desafio - Sistema de Triagem de Candidatos**

O sistema representa cada candidato como um dicionário com a seguinte estrutura:

```python
{
    "nome": "Ana Lima",
    "cargo_desejado": "Analista de Dados",
```

### Parte A — Representar um candidato

**Função 1:** `criar_candidato(nome, cargo, anos, pretensao, modalidade)`

Retorna um dicionário no formato descrito acima.

**Exemplo:**

```
criar_candidato("Ana Lima", "Analista de Dados", 1, 55000, "remoto")
->  {"nome": "Ana Lima", "cargo_desejado": "Analista de Dados",
     "anos_experiencia": 1, "pretensao_salarial": 55000, "modalidade": "remoto"}
```



**Função 2:** `apresentar_candidato(candidato)`

Recebe um dicionário de candidato e retorna uma string formatada.

**Exemplo de saída:**

```
Ana Lima
  Cargo desejado:  Analista de Dados
  Nível:           Júnior (1 ano de experiência)
  Pretensão:       USD 55,000/ano
  Modalidade:      remoto
```


**Função 3:** `esta_na_faixa(candidato, salario_min, salario_max)`

Retorna `True` se a pretensão salarial estiver no intervalo (inclusive nos extremos), e `False` caso contrário.

**Exemplo:**

```
ana = criar_candidato("Ana Lima", "Analista", 1, 55000, "remoto")
esta_na_faixa(ana, 50000, 60000)  ->  True
esta_na_faixa(ana, 60000, 80000)  ->  False
```

````python
def criar_candidato(nome, cargo, anos, pretensao, modalidade):

    # sua solução aqui
    candidato = {
        "nome": nome,
        "cargo_desejado": cargo,
        "anos_experiencia": anos,
        "pretensao_salarial": pretensao,
        "modalidade": modalidade
    }
    return candidato


def apresentar_candidato(candidato):
    # sua solução aqui
    return f"{candidato['nome']}\n  Cargo desejado:  {candidato['cargo_desejado']}\n  Nível:           {classificar_nivel(candidato['anos_experiencia'])} ({candidato['anos_experiencia']} anos de experiência)\n  Pretensão:       USD {candidato['pretensao_salarial']}/ano\n  Modalidade:      {candidato['modalidade']}"


def esta_na_faixa(candidato, salario_min, salario_max):
    # sua solução aqui
    esta_na_faixa = False
    if candidato['pretensao_salarial'] >= salario_min and candidato['pretensao_salarial'] <= salario_max:
      esta_na_faixa = True
    return esta_na_faixa
````

````python
# Testes da Parte A
ana = criar_candidato("Ana Lima", "Analista de Dados", 1, 55000, "remoto")

print(ana["nome"] == "Ana Lima")
print(ana["anos_experiencia"] == 1)
print(ana["pretensao_salarial"] == 55000)

print(esta_na_faixa(ana, 50000, 60000) == True)
print(esta_na_faixa(ana, 60000, 80000) == False)
print(esta_na_faixa(ana, 55000, 55000) == True)

# Inspecione a saída visualmente
print()
print(apresentar_candidato(ana))
````

### Parte C - Análise do banco

**Função 7:** `media_pretensao(banco)`

Retorna a média das pretensões salariais, arredondada para 2 casas decimais.

**Exemplo:**

```
media_pretensao(banco)  ->  91000.0
```



**Função 8:** `candidatos_na_faixa(banco, salario_min, salario_max)`

Retorna uma lista com os dicionários dos candidatos cuja pretensão está na faixa informada.

**Exemplo:**

```
candidatos_na_faixa(banco, 55000, 100000)  ->  [<lista com 4 candidatos>]
```



**Função 9:** `ranking_por_experiencia(banco)`

Retorna a lista de candidatos ordenada do mais experiente para o menos experiente.

**Exemplo:**

```
ranking = ranking_por_experiencia(banco)
ranking[0]["nome"]   ->  "Carla Mendes"  (8 anos)
ranking[-1]["nome"]  ->  "Ana Lima"      (1 ano)
```

````python
def media_pretensao(banco):
    # sua solução aqui
    if not banco:
        return 0.0
    total_pretensao = sum(c['pretensao_salarial'] for c in banco)
    media = total_pretensao / len(banco)
    return round(media, 2)


def candidatos_na_faixa(banco, salario_min, salario_max):
    # sua solução aqui
    return [c for c in banco if esta_na_faixa(c, salario_min, salario_max)]


def ranking_por_experiencia(banco):
    # sua solução aqui
    return sorted(banco, key=lambda x: x['anos_experiencia'], reverse=True)
````

````python
# Testes da Parte C — use o mesmo banco da Parte B
media_esperada = round((55000 + 95000 + 145000 + 60000 + 100000) / 5, 2)
print(media_pretensao(banco) == media_esperada)

print(len(candidatos_na_faixa(banco, 55000, 100000)) == 4)
print(len(candidatos_na_faixa(banco, 0, 50000)) == 0)

ranking = ranking_por_experiencia(banco)
print(ranking[0]["nome"]  == "Carla Mendes")
print(ranking[-1]["nome"] == "Ana Lima")
````

````python
def validar_dados_candidato(dados):
    chaves_obrigatorias = ["nome", "cargo_desejado", "anos_experiencia", "pretensao_salarial", "modalidade"]
    modalidades_validas = ["remoto", "presencial", "híbrido"]

    for chave in chaves_obrigatorias:
        if chave not in dados:
            raise ValueError(f"Chave faltando: {chave}")

    if dados["anos_experiencia"] < 0:
        raise ValueError(f"Anos de experiência inválidos: {dados['anos_experiencia']}")

    if dados["pretensao_salarial"] < 0:
        raise ValueError(f"Pretensão salarial inválida: {dados['pretensao_salarial']}")

    if dados["modalidade"] not in modalidades_validas:
        raise ValueError(f"Modalidade inválida: {dados['modalidade']}")

    return True
````
